```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

path_to_data = '/home/fraga/Documents/Series-Temporales/Data/airline_passengers.csv'
```

```python
airline = pd.read_csv(path_to_data, index_col = 0, parse_dates = True)
airline.shape
```

```python
airline.head()
```

```python
airline.plot()
plt.show()
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'], model = 'additive')
result.plot()
plt.show()
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'], model = 'multiplicative')
result.plot()
plt.show()
```
