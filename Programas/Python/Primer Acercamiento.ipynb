```python
import scipy.stats
import pandas as pd
import matplotlib.pyplot as plt

path_to_data = '/home/fraga/Documents/Series Temporales/Data/Index2018.csv'
```

## Reading the data from a .csv file with pandas

```python
raw_csv_data = pd.read_csv(path_to_data)
```

```python
raw_csv_data.spx.plot(figsize = (25, 5))
plt.title('S&P500 Prices', size = 20)
plt.show()
```

```python
raw_csv_data.ftse.plot(figsize = (25, 5))
plt.title('FTSE100 Prices', size = 20)
plt.show()
```

```python
raw_csv_data.spx.plot(figsize = (20, 5), title = 'S&P500 Prices')
raw_csv_data.ftse.plot(figsize = (20, 5), title = 'FTSE100 Prices')
plt.title("S&P500 vs FTSE100", size = 20)
plt.show()
```

```python
scipy.stats.probplot(raw_csv_data.spx, plot = plt)
plt.title("QQ Plot", size = 20)
plt.show()
```
